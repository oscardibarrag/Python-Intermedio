```python
titulos = []
titulos_items = soup.find_all('h3') #Trae las etiqueta HTML <h3> que encuentra en la pagina y aparte traiga un atributo llamado title, para traer los titulos de los libros

for items in titulos_items:
    titulo = items.a
    #print(titulo['title']) <-- imprimir directo los titulos
    titulos.append(titulo['title']) #<-- Guardar los titulos en un arreglo
    
    
```

```python
precios = []

precios_items = soup.find_all('p', class_= "price_color")

for items in precios_items:
    precio = items.text
    precios.append(precio)

precios
```

```python
stocks = []

stocks_items = soup.find_all('p', class_='instock availability')

for item in stocks_items:
    stock = item.text
    stocks.append(stock)

stocks
```

```python
ratings = []

def extraer_ratings (soup):
    page_items = soup.find_all('p', class_='star-rating')
    
    for item in page_items:
        ratings.append(item['class'][1]) #Extraer el segundo valor del atributo class 
        
extraer_ratings(soup)
ratings
```

## Paginación  

```python
import requests

from bs4 import BeautifulSoup
```

```python
def obtener_contenido_pagina(url):
    response = requests.get(url)
    return response.content
```

```python
def analizar_contenido_html(html):
    return BeautifulSoup(html, 'html.parser')
```

```python
data = []

def procesar_pagina(soup):
    titulos = []
    links = []
    ratings = []
    existencias = []

    titulo_items = soup.find_all('h3')

    for items in titulo_items:
        titulo = items.a
        titulos.append(titulo['title'])
        links.append(titulo['href'])

    rating_items = soup.find_all('p', class_="star-rating")

    for item in rating_items:
        rating = item['class'][1]
        ratings.append(rating)

    existencia_items = soup.find_all('p', class_='instock availability')

    for item in existencia_items:
        existencia = item.text.strip()
        existencias.append(existencia)

    for i in range(len(titulos)):
        data.append({
            "Título": titulos[i],
            "Rating": ratings[i],
            "Link": links[i],
            "Existencia": existencias[i]
        })
```

```python
def manejar_paginacion(url_base, num_paginas):
    for i in range(1, num_paginas + 1):
        url = url_base + '/page-' + str(i) + ".html"  # Actualiza la URL base con el número de página actual
        contenido_pagina = obtener_contenido_pagina(url)
        soup = analizar_contenido_html(contenido_pagina)
        procesar_pagina(soup)
```

```python
import re

texto = "Hola, mi número de teléfono es 123-456-7890. ¿Necesitas contactarme? Llámame. Contacto: 123-456-7897"

patron = r'\d{3}-\d{3}-\d{4}'

resultado_search = re.search(patron, texto)

if resultado_search:
    print("Número de teléfono encontrado:", resultado_search.group())
else:
    print("Número de teléfono no encontrado.")
```

```python
patron_saludo = r'Hola'

resultado_match = re.match(patron_saludo, texto)
    
if resultado_match:
    print("Saludo encontrado:", resultado_match.group())
else:
    print("Saludo no encontrado.")
```

```python
patron_telefonos = r'\d{3}-\d{3}-\d{4}'
resultado_findall = re.findall(patron_telefonos, texto)

if resultado_findall:
    print("Números de teléfono encontrados:", resultado_findall)
else:
    print("Ningún número de teléfono encontrado.")
```
