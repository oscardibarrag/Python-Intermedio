```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def obtener_contenido_pagina(url):
    response = requests.get(url)
    return response.text
```

```python
def analizar_contenido_html(html):
    #print(html)
    return BeautifulSoup(html, 'html.parser')
```

```python
data =[]
def procesar_pagina(soup):
    Titulo = []
    Autor = []
    Fecha = []
    Link = []
    Introduccion = []
    
    titulo_items = soup.find_all('h2', class_="title")
    
    for item in titulo_items:
        titulos = item.a.text.strip()
        Titulo.append(titulos)
     
    autor_items = soup.find_all('div', {'class':'meta'})
    
    for item in autor_items:
        elemento_a = item.find('a')
        if elemento_a:       
            autores=elemento_a.text.strip()
            Autor.append(autores)

    fecha_items =  soup.find_all('div', class_="meta")
    
    for item in fecha_items:
        elemento_span = item.find_all('span')
        if len(elemento_span) >= 3:
            fechas = elemento_span[2].text.strip()
            Fecha.append(fechas)
    
    link_items = soup.find_all('h2', class_="title")
    
    for item in link_items:
        links = item.a['href']
        Link.append(links)
        
    intro_items = soup.find_all('div', class_="excerpt")

    for item in intro_items:
        intros = item.text.strip()
        Introduccion.append(intros)
    
    
    min_length = min(len(Titulo), len(Autor), len(Fecha), len(Introduccion))
    for i in range(min_length):
        data.append({
            "Titulo": Titulo[i] if i < len(Titulo) else None,
            "Autor": Autor[i] if i < len(Autor) else None,
            "Fecha": Fecha[i] if i < len(Fecha) else None,
            "Link": Link[i] if i < len(Link) else None,
            "Introduccion": Introduccion[i] if i < len(Introduccion) else None,
            
        })
```

```python
def manejar_paginacion(url_base, num_paginas):
    for i in range(1, num_paginas + 1):
            url = url_base + str(i)+'/'
            contenido_pagina = obtener_contenido_pagina(url)
            soup = analizar_contenido_html(contenido_pagina)
            procesar_pagina(soup)

```

```python
url_base = 'https://laparadadigital.com/category/destacadas/page/'
num_paginas = 30
manejar_paginacion(url_base, num_paginas)
```

```python
import pandas as pd
df = pd.DataFrame(data)
```

```python
import datetime

fecha_actual = datetime.datetime.now().strftime("%d-%m-%Y")
df.to_csv(f"dataset/web scraping/Noticias_destacadas_historico-{fecha_actual}.csv", index=False)
```
