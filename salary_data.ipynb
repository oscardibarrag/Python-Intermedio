```python
import pandas as pd

datos = pd.read_csv('Salary_data.csv')
datos
```

```python
datos.info()
```

```python
#Revisar cuales columnas tienen campos vacios

datos.isnull().sum()
```

```python
#Remplazar valores vacios

media = datos["Age"].median()
datos_age = datos["Age"].fillna(media)

vacio = 'Sin registro'
datos_gender = datos['Gender'].fillna(vacio)
datos_education = datos['Education Level'].fillna(vacio)
datos_job = datos['Job Title'].fillna(vacio)

media = datos['Years of Experience'].median()
datos_experience = datos['Years of Experience'].fillna(media)

media = datos['Salary'].median()
datos_salary = datos['Salary'].fillna(media)

datos_sv = pd.DataFrame({
    'Age':datos_age,
    'Gender':datos_gender,
    'Education Level':datos_education,
    'Job title':datos_job,
    'Years of Experience':datos_experience,
    'Salary':datos_salary})



```

```python
#Datos corregidos
datos_sv.info()
```

```python


unique_count = datos_sv['Job title'].nunique()
unique_values = datos_sv['Job title'].unique()
print(f"Columna -Job Title- tiene {unique_count} valores unicos")
#print(unique_values)

maximo = datos_sv['Job title'].max()
print(f"El Titulo de trabajo con mejor salario es: {maximo}")
minimo = datos_sv['Job title'].min()
print(f"El Titulo de trabajo con menor salario es: {minimo}")



```

```python
import matplotlib as plt
import matplotlib.pyplot as plt

font1 = {'family':'serif','color':'blue','size':15}
font2 = {'family':'serif','color':'red','size':20}


x = datos_sv['Age']
y = datos_sv['Salary']

#plt.subplot(1,2,1)
plt.scatter(x,y)
plt.xlabel("Edad",fontdict=font1)
plt.ylabel("Salario",fontdict=font1)
plt.title("RelaciÃ³n  Edad - Salario", fontdict=font2)


plt.show()
```
